```python
from IPython.display import display, Math
import sympy as sym
import re
import dataclasses

class trajectory:
    pass

sym.init_printing()

def calculate_trajectory(l,show_full=False ):

   

    raise_to_pow_2_latex = lambda e: f'2^{e}'


    raise_to_pow_2 = lambda e: 2**(e)
    
    exponent =next(l,0)
    
        
    num = raise_to_pow_2(exponent)

    latex = raise_to_pow_2_latex(exponent)

    for exponent in l:
        # print( num )
        if num % 3 !=1:
            raise ValueError("The trajectory is leading to a non integer!")
        

        num = ((num -1) // 3) * raise_to_pow_2(exponent)

        latex = r"\frac{" +latex+ "-1}{3}"
        if(exponent>0):
            latex = latex + "*{"+raise_to_pow_2_latex(exponent)+"}"

    display(Math(latex))
    return num
```
